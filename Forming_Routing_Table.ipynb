```
import json
import os

def record_bgp(file_j,file_f):
  if os.path.exists(file_f):
    f=open(file_f,'r')
    data_f=json.load(f)
  else:
    data_f={}
  f=open(file_j,'r')
  data=json.load(f)
  

  for key in data.keys():
    if 'FROM' not in data[key].keys() or 'TO' not in data[key].keys():
      continue
    else:
      from_v=data[key]['FROM']
      if data_f.get(from_v) is not None:
        to_v=data[key]['TO']
        if data_f[from_v].get(to_v) is not None:
          Time_v=data[key]['TIME']
          data_f[from_v][to_v][Time_v]={}
          if "ASPATH" in data[key].keys():
            as_p=data[key].get("ASPATH").split(" ")

            data_f[from_v][to_v][Time_v]["AS_PATH"]=as_p
          else:
            data_f[from_v][to_v][Time_v]["AS_PATH"]=["N/A"]
          if "ANNOUNCE" in data[key].keys():
            as_p=data[key].get("ANNOUNCE")

            data_f[from_v][to_v][Time_v]["ANNOUNCE"]=as_p
          else:
            data_f[from_v][to_v][Time_v]["ANNOUNCE"]=["N/A"]
          if "WITHDRAW" in data[key].keys():
            as_p=data[key].get("WITHDRAW")

            data_f[from_v][to_v][Time_v]["WITHDRAW"]=as_p
          else:
            data_f[from_v][to_v][Time_v]["WITHDRAW"]=["N/A"]
          if "NEXT_HOP" in data[key].keys():
            as_p=data[key].get("NEXT_HOP")

            data_f[from_v][to_v][Time_v]["NEXT_HOP"]=as_p
          else:
            data_f[from_v][to_v][Time_v]["NEXT_HOP"]="N/A"
        else:
          data_f[from_v][to_v]={}
          Time_v=data[key]['TIME']
          data_f[from_v][to_v][Time_v]={}
          if "ASPATH" in data[key].keys():
            as_p=data[key].get("ASPATH").split(" ")

            data_f[from_v][to_v][Time_v]["AS_PATH"]=as_p
          else:
            data_f[from_v][to_v][Time_v]["AS_PATH"]=["N/A"]
          if "ANNOUNCE" in data[key].keys():
            as_p=data[key].get("ANNOUNCE")

            data_f[from_v][to_v][Time_v]["ANNOUNCE"]=as_p
          else:
            data_f[from_v][to_v][Time_v]["ANNOUNCE"]=["N/A"]
          if "WITHDRAW" in data[key].keys():
            as_p=data[key].get("WITHDRAW")

            data_f[from_v][to_v][Time_v]["WITHDRAW"]=as_p
          else:
            data_f[from_v][to_v][Time_v]["WITHDRAW"]=["N/A"]
          if "NEXT_HOP" in data[key].keys():
            as_p=data[key].get("NEXT_HOP")

            data_f[from_v][to_v][Time_v]["NEXT_HOP"]=as_p
          else:
            data_f[from_v][to_v][Time_v]["NEXT_HOP"]="N/A"
      else:
        data_f[from_v]={}
        to_v=data[key]['TO']
        data_f[from_v][to_v]={}
        Time_v=data[key]['TIME']
        data_f[from_v][to_v][Time_v]={}
        if "ASPATH" in data[key].keys():
          as_p=data[key].get("ASPATH").split(" ")

          data_f[from_v][to_v][Time_v]["AS_PATH"]=as_p
        else:
          data_f[from_v][to_v][Time_v]["AS_PATH"]=["N/A"]
        if "ANNOUNCE" in data[key].keys():
            as_p=data[key].get("ANNOUNCE")

            data_f[from_v][to_v][Time_v]["ANNOUNCE"]=as_p
        else:
            data_f[from_v][to_v][Time_v]["ANNOUNCE"]=["N/A"]
        if "WITHDRAW" in data[key].keys():
            as_p=data[key].get("WITHDRAW")

            data_f[from_v][to_v][Time_v]["WITHDRAW"]=as_p
        else:
            data_f[from_v][to_v][Time_v]["WITHDRAW"]=["N/A"]
        if "NEXT_HOP" in data[key].keys():
            as_p=data[key].get("NEXT_HOP")

            data_f[from_v][to_v][Time_v]["NEXT_HOP"]=as_p
        else:
            data_f[from_v][to_v][Time_v]["NEXT_HOP"]="N/A"
  print(len(data_f))
  print(data_f)   
  print(file_f)
  with open(file_f, "w") as outfile: 
        json.dump(data_f, outfile) 
          





    
```

```
import pandas as pd

def BGP_to_csv(file_f,file_c):
  
 
  
  Announce_l=[]
  Withdraw_l=[]
  AS_l=[]
  text=open(file_c,'w')
  text.write('From,To,Date,Time,Next_Hop\n')
  text.close()
  text=open(file_c,'a')
  f=open(file_f,'r')
  data=json.load(f)
  for key_1 in data.keys():
    for key_2 in data[key_1].keys():
      for key_3 in data[key_1][key_2].keys():
        text.write(key_1.split(" ")[1]+','+key_2.split(" ")[1]+','+key_3.split(" ")[0]+','+key_3.split(" ")[1]+','+data[key_1][key_2][key_3]["NEXT_HOP"]+'\n')
        AS_l.append(data[key_1][key_2][key_3]["AS_PATH"])
        Announce_l.append(data[key_1][key_2][key_3]["ANNOUNCE"])
        Withdraw_l.append(data[key_1][key_2][key_3]["WITHDRAW"])
  text.close()
  #print(file_c)
  df=pd.read_csv(file_c,encoding='utf-8',engine='python')
  #df_1=pd.DataFrame(df)
  #print(df.head())
  #print(type(df))
  print(type(AS_l))

  print(type(Announce_l))
  df['AS_PATH']=AS_l
  df['Announce']=Announce_l
  df['Withdraw']=Withdraw_l
  print("A")
  print(df.head())
  df.to_csv(file_c,index=False)


  


```

```
def caller():
   Path="drive/My Drive/iitkgp_summer_project/"
   ext='Routing_data'
   ext_2='csv_converted_det'
   file_1=Path+ext
   file_2=Path+ext_2
   files=os.listdir(file_1)
   for fle in files:
    ext_main=fle[:-5]+'.csv'
    file_k=file_1+'/'+fle
    file_c=file_2+'/'+ext_main
    BGP_to_csv(file_k,file_c)
```
